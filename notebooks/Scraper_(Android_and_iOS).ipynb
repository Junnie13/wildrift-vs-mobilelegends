* scraping and cleaning from google playstore

```python
# reviews_all() method extracts ALL of the reviews from google playstore, the others are the parameters for the reviews_all() method (more here: https://pypi.org/project/google-play-scraper/)
google_reviews, continuation_token = reviews(
        "com.globe.gcash.android", # https://play.google.com/store/apps/details?id=com.globe.gcash.android&hl=en&gl=US&pli=1
        lang = 'en', # defaults to 'en'
        country = 'us', # defaults to 'us'
        sort = Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
        count = 10000 # default = 100
    )

result, _ = reviews(
    'com.riotgames.league.wildrift',
    continuation_token= continuation_token # defaults to None (load from the beginning)
)

# dataframe that takes all of the reviews from google play
google_df = pd.DataFrame(np.array(google_reviews), columns=['review'])
google_df2 = google_df.join(pd.DataFrame(google_df.pop('review').tolist()))

google_df2
```

```python
# reviews_all() method extracts ALL of the reviews from google playstore, the others are the parameters for the reviews_all() method (more here: https://pypi.org/project/google-play-scraper/)
google_reviews = reviews_all(
        "com.globe.gcash.android", # https://play.google.com/store/apps/details?id=com.globe.gcash.android&hl=en&gl=US&pli=1
        lang = 'en', # defaults to 'en'
        country = 'us', # defaults to 'us'
        sort = Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
        # count = 10000 # default = 100
    )

# result, _ = reviews(
#     'com.riotgames.league.wildrift',
#     continuation_token= continuation_token # defaults to None (load from the beginning)
# )

# dataframe that takes all of the reviews from google play
google_df = pd.DataFrame(np.array(google_reviews), columns=['review'])
google_df2 = google_df.join(pd.DataFrame(google_df.pop('review').tolist()))

google_df2
```

```python
# dropping irrelevant columns from list
google_df2.drop(columns={'reviewId', 'userName', 'userImage', 'reviewCreatedVersion'},inplace = True)
#renaming columns:
google_df2.rename(columns= {'score': 'rating', 'content': 'review', 'at': 'date', 'replyContent': 'developerResponse', 'repliedAt': 'developerResponse_date', 'thumbsUpCount': 'thumbs_up'},inplace = True)
google_df2.insert(loc=0, column='source', value='Google Play')
google_df2.insert(loc=3, column='review_title', value=None)
# google_df2['language_code'] = 'en'
# google_df2['country_code'] = 'us'

google_df2
```

```python
google_df2.to_csv("gcash_playstore.csv")
```
